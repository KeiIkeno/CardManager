```python
import pandas as pd
```

```python
import urllib
from bs4 import BeautifulSoup
```

```python
import time
```

```python
urls = 'https://ocg.xpg.jp/list/list.fcgi?Set=554
```

```python
summary = pd.DataFrame()
```

```python
for url in urls[0][120:]:
    fp = urllib.request.urlopen(url)
    soup = BeautifulSoup(fp, "html.parser")
    for body in soup("tbody"):
        body.unwrap()

    df = pd.read_html(str(soup), flavor="bs4", header=0)
    
    summary = summary.append(df)
    time.sleep(5)
```

```python
summary
```

```python
summary.to_csv('cardlist_new.csv')
```
